```python
import pandas as pd

# CSVファイルを読み込む
df = pd.read_csv('outputA_2.csv')

df.drop(columns=['Question_Num', 'url', 'classifier', 'WA_count'], inplace=True)

# ユニークな単語を保持するセット
unique_ja_words = set()
unique_en_words = set()

# ja_textおよびen_textのすべてのユニークな単語を集める
for ja_text, en_text in zip(df['ja_text'], df['en_text']):
    if isinstance(ja_text, str):
        unique_ja_words.update(ja_text.split(','))
    if isinstance(en_text, str):
        unique_en_words.update(en_text.split(','))

# 各ユニーク単語の出現回数を計算する
for word in unique_ja_words:
    df[word] = df['ja_text'].apply(lambda x: x.split(',').count(word) if isinstance(x, str) else 0)

for word in unique_en_words:
    df[word] = df['en_text'].apply(lambda x: x.split(',').count(word) if isinstance(x, str) else 0)

df.drop(columns=['ja_text', 'en_text'], inplace=True)

# 結果を新しいCSVファイルとして保存
df.to_csv('outputA_3.csv', index=False, sep=';')
```
